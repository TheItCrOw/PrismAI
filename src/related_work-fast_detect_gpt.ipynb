```python
import torch

from transition_scores.data import FeatureValues


def get_sampling_discrepancies_from_probs(
    likelihood: torch.Tensor,
    sampled_probs: torch.Tensor,
    full_probs: torch.Tensor | None = None,
) -> float:
    log_likelihood = likelihood.log()
    log_sampled_probs = (
        full_probs.log() if full_probs is not None else sampled_probs.log()
    )
    mean_ref = (sampled_probs * log_sampled_probs).sum(dim=-1)
    var_ref = (sampled_probs * log_sampled_probs.square()).sum(
        dim=-1
    ) - mean_ref.square()
    discrepancy = (log_likelihood.sum(dim=-1) - mean_ref.sum(dim=-1)) / var_ref.sum(
        dim=-1
    ).sqrt()
    return discrepancy.mean().item()


def get_sampling_discrepancies_from_features(ts: FeatureValues) -> float:
    return get_sampling_discrepancies_from_probs(
        torch.tensor(ts.target_probs),
        torch.tensor(ts.top_k_probs),
    )
```
